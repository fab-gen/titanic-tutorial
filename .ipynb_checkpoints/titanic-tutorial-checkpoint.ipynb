## **Age distributions**

The missing values of the Age variable with be filled with the median, which isn't sensitive to outliers. But using the median of all the dataset is not the wisest choice so let's explore if we can find more appropriate groups.

Since Age is correlated with Pclass, let's plot the Age distribution for each class.

#### *Ages by class plotting*

```python
fig, ax = plt.subplots(figsize=(10, 5))

colors = ["g", "b", "r"]

for cls in range(1, 4):

    color = colors[cls - 1]
    class_data = train_data[(train_data["Pclass"] == cls)]

    sns.kdeplot(class_data, x="Age", color=color, ax=ax,
                clip=(0, max(train_data["Age"])))

    age_mean = class_data["Age"].mean()
    age_median = class_data["Age"].median()

    plt.axvline(x=age_mean, color=color, linestyle="-",
                label="Class {} Mean = {:.1f}".format(cls, age_mean))

    plt.axvline(x=age_median, color=color, linestyle="--",
                label="Class {} Median = {:.1f}".format(cls, age_median))


plt.xticks(np.arange(0, 100, 5))

plt.tight_layout()

plt.legend()

plt.plot();
```

We can notice that the younger people are, the more likely they are to have a Class 3 ticket. On the opposite, the older people are, the more likely they are to have a Class 1 ticket.

Let's also plot the Age distribution for each Sex.

```python
fig, ax = plt.subplots(figsize=(10, 5))

colors = ["violet", "darkviolet"]
sex_list = ["female", "male"]


for sex_int in range(2):

    sex = sex_list[sex_int]
    color = colors[sex_int]
    sex_data = train_data[(train_data["Sex"] == sex)]

    sns.kdeplot(sex_data, x="Age", color=color, ax=ax,
                clip=(0, max(train_data["Age"])))

    age_mean = sex_data["Age"].mean()
    age_median = sex_data["Age"].median()

    plt.axvline(x=age_mean, color=color, linestyle="-",
                label="{} Mean = {:.1f}".format(sex, age_mean))

    plt.axvline(x=age_median, color=color, linestyle="--",
                label="{} Median = {:.1f}".format(sex, age_median))


plt.xticks(np.arange(0, 100, 5))

plt.tight_layout()

plt.legend()

plt.plot();
```

We notice that the means and the medians are pretty close but still different. We notice also that the women are a little younger than the men.

## **Data Cleaning**

### **Age missing values**

Since the ages differ according to the class and the sex, it will be wise to group the ages according to these two features and to fill null values with group medians.

```python
median_ages = pd.DataFrame(train_data.groupby(["Pclass", "Sex"])["Age"].median()).reset_index()
median_ages
```

```python
for df in [train_data, test_data]:
    merged_df = df.merge(median_ages, on=["Pclass", "Sex"], how="left", suffixes=("", "_median"), copy=False)
    merged_df["Age"].fillna(merged_df["Age_median"], inplace=True)
    merged_df.drop(columns=["Age_median"], inplace=True)
    df.update(merged_df)
```

## **Feature Engineering**

### **Binning**

The main goal of binning is to transform numeric features in categorical features. With one-hot-encoding, we have as much as features as categories. This gives more choices to the greedy approach of the decision tree to select the best feature at each step.

### **Fare binning**

In order to correctly apply the binnings to the test dataset, we have to check if the range of the training dataset includes the range of the test dataset.

```python
print(min(train_data["Fare"]), max(train_data["Fare"]))
```

```python
print(min(test_data["Fare"]), max(test_data["Fare"]))
```

OK, we're good on that.

Viewing the histograms of the Fare feature, a range from 11 to 16 seems a good start for the number of quantiles for the cut.

```python
for q in range(11, 17):
    train_data['Fare {}Q'.format(q)] = pd.qcut(train_data['Fare'], q, precision=4)
```

Let's plot the different numbers of quantiles.

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(20, 35))

for i in range (1, 7):
    q = 17 - i # to get the biggest number of quantiles (16) in first spot
    feature = "Fare {}Q".format(q)
    plt.subplot(3, 2, i)
    sns.countplot(data=train_data, x=feature, hue="Survived")
    plt.xlabel("{}".format(feature))
    plt.xticks(rotation=45)
    plt.legend(["Not Survived", "Survived"])

plt.show()
```

Viewing these different plots, 14 seems a good trade-off between preserving quantiles with different survival rates and having quantiles with a not too small range.

Let's replace the Fare values with the quantiles.

```python
for q in range(11, 17):
    if q == 14:
        train_data["Fare"] = train_data["Fare {}Q".format(q)]
    train_data.drop("Fare {}Q".format(q), axis=1, inplace=True)
```

```python
train_data.head()
```

And now, let's apply the cut to the test dataset.

```python
train_fare_intervals = pd.IntervalIndex(train_data["Fare"]).unique()
```

```python
test_data["Fare"] = test_data["Fare"].apply(lambda x: train_fare_intervals[train_fare_intervals.get_loc(x)]).astype("category")
```

```python
test_data.head()
```

### **Age binning**

Let's do the same for the Age feature.

Let's check the range of the training dataset and the test dataset.

```python
print(min(train_data["Age"]), max(train_data["Age"]))
```

```python
print(min(test_data["Age"]), max(test_data["Age"]))
```

We notice that the minimum age of the test dataset is below the minimum age of the training dataset.

To solve this issue, let's round the ages below 1, it won't do much difference.

```python
train_data["Age"] = train_data["Age"].apply(lambda x: round(x) if x < 1 else x)
```

```python
test_data["Age"] = test_data["Age"].apply(lambda x: round(x) if x < 1 else x)
```

Viewing the histograms of the Age feature, a range from 6 to 11 seems a good start for the number of quantiles with which we will cut the Age values.

```python
for q in range(6, 12):
    train_data["Age {}Q".format(q)] = pd.qcut(train_data["Age"], q, duplicates="drop")
```

Let's plot the different numbers of quantiles.

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(20, 30))

for i in range (1, 7):
    q = 12 - i # to get the biggest number of quantiles (11) in first spot
    feature = "Age {}Q".format(q)
    plt.subplot(3, 2, i)
    sns.countplot(data=train_data, x=feature, hue="Survived")
    plt.xlabel("{}".format(feature))
    plt.xticks(rotation=45)
    plt.legend(["Not Survived", "Survived"])

plt.show()
```

Viewing these different plots, 11 seems a good choice.

Let's replace the Age values with the quantiles.

```python
for q in range(6, 12):
    if q == 11:
        train_data["Age"] = train_data["Age {}Q".format(q)]
    train_data.drop("Age {}Q".format(q), axis=1, inplace=True)
```

```python
train_data.head()
```

And now, let's apply the cut to the test dataset.

```python
train_age_intervals = pd.IntervalIndex(train_data["Age"]).unique()
```

```python
test_data["Age"] = test_data["Age"].apply(lambda x: train_age_intervals[train_age_intervals.get_loc(x)]).astype("category")
```

```python
test_data.head()
```

The family type is still a strong predictor.

Another important feature to consider is the Title of each passenger, which is at the beggining of each passenger's name. It indicates the social position of each passenger and can be useful to determine the survival.

Let's extract this new feature.

```python
train_data["Title"] = train_data["Name"].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

plt.subplot(1, 2, 1)
sns.countplot(data=train_data, x="Title", ax=axs[0])
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.countplot(data=train_data, x="Title", ax=axs[1], hue="Survived")
plt.xticks(rotation=45)
plt.legend(["Not Survived", "Survived"]);
```

By looking at the plots, we can see that we could exploit this feature to help us predict the survival rate. But some titles regard few passengers. Maybe we could group some of these features in order to reduce their number.

Let's see the value counts for the training and the test datasets.

```python
train_data["Title"].value_counts()
```

```python
test_data["Title"].value_counts()
```

If we want to do the same grouping between the training and the test datasets, we have to check if the training dataset includes all the titles of the test dataset.

```python
set(test_data["Title"]) - set(train_data["Title"])
```

The title "Dona" is not in the training dataset, so we will have to add it to our group list.

The titles "Dr", "Col", "Major", "Jonkheer", "Capt", "Sir", "Don", "Rev", "Lady", "Dona", "the Countess" regard few passengers and are noble or professional titles. It should be useful to group them in a "Dr/Military/Noble/Clergy" feature.

```python
train_data["Title"].replace(["Dr", "Col", "Major", "Jonkheer", "Capt", "Sir", "Don", "Rev", "Lady", "Dona", "the Countess"], "Dr/Military/Noble/Clergy", inplace=True)
test_data["Title"].replace(["Dr", "Col", "Major", "Jonkheer", "Capt", "Sir", "Don", "Rev", "Lady", "Dona", "the Countess"], "Dr/Military/Noble/Clergy", inplace=True)
```

The titles "Miss", "Mrs", "Ms", "Mlle", "Mme" regard women and their marital status. It should be useful to group these titles relatively to their marital situation.

```python
train_data["Title"].replace(["Mrs", "Ms", "Mme"], "Mrs", inplace=True)
train_data["Title"].replace(["Miss", "Mlle"], "Miss", inplace=True)

test_data["Title"].replace(["Mrs", "Ms", "Mme"], "Mrs", inplace=True)
test_data["Title"].replace(["Miss", "Mlle"], "Miss", inplace=True)
```

Let's plot the distributions of this new feature.

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

plt.subplot(1, 2, 1)
sns.countplot(data=train_data, x="Title", ax=axs[0])
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.countplot(data=train_data, x="Title", ax=axs[1], hue="Survived")
plt.xticks(rotation=45)
plt.legend(["Not Survived", "Survived"]);
```

By looking at the plots, we notice that this feature could be useful to predict the survival rate.

Let's also see the features importance.

```python
def plot_features_importances(clf):
    
    feat_importances_df = pd.DataFrame(data=clf.feature_importances_, 
                                   index=feature_names, 
                                   columns=["Importance"])

    feat_importances_df = feat_importances_df.sort_values(by="Importance").reset_index(names="Feature")
    
    plt.figure(figsize=(8, 10))
    plt.barh(feat_importances_df["Feature"], feat_importances_df["Importance"], align="center")
    plt.xlabel("Feature importance")
    plt.ylabel("Feature");
```

```python
plot_features_importances(dt_clf)
```

Since the max depth of our decision tree is not very high, the number of used features is limited. The feature "Title_Mr" is the most important feature, because it is used at the top node of the decision tree. Thus, it is the feature which regards the most samples and whose node creates the best impurity decrease.

### **XG Boost**

Now let's us Gradient Boosting Decision Trees, which are an even more efficient version of decision trees. This algorithm works with "weak learners" which are very shallow decision trees. The first decision tree is fitted to the training dataset and the next ones are fitted to the errors of the previous one.

Since we choose a high number of estimators (1000), we compensate it by choosing a small value for the learning rate (0.01).

```python
xgb_clf = XGBClassifier(n_jobs=-1, n_estimators=1000, learning_rate=0.01, max_depth=1)
```

```python
xgb_clf.fit(X, y)
```
